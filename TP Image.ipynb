# Seuillage et binarisation

```python
#binarisation d'une image en couleur MÃ©thode 1
def binarisationC1(F,p):
    nimg=np.copy(F)
    (largeur,hauteur,ncouleur)=F.shape
    for i in range(largeur):
        for j in range(hauteur):
            a=0.33*nimg[i][j][0]+0.33*nimg[i][j][2]+0.33*nimg[i][j][2]
            for k in range(3):
                if a>p:
                    nimg[i][j][k]=1
                else :
                    nimg[i][j][k]=0
    plt.imshow(nimg,cmap="gray")
    plt.show()
```

```python
#binarisation d'une image en couleur MÃ©thode 2
def binarisationC2(F,p1,p2,p3):
    nimg=np.copy(F)
    (largeur,hauteur,ncouleur)=F.shape
    for i in range(largeur):
        for j in range(hauteur):
                if nimg[i][j][0]>p1:
                    nimg[i][j][0]=1
                else :
                    nimg[i][j][0]=0
    for i in range(largeur):
        for j in range(hauteur):
                if nimg[i][j][1]>p2:
                    nimg[i][j][1]=1
                else :
                    nimg[i][j][1]=0
    for i in range(largeur):
        for j in range(hauteur):
                if nimg[i][j][2]>p3:
                    nimg[i][j][2]=1
                else :
                    nimg[i][j][2]=0
    plt.imshow(nimg,cmap="gray")
    plt.show()
```

```python
#Seuillage d'une image au niveau de gris
def Seuillage(F,p):
    nimg=np.copy(F)
    (largeur,hauteur)=F.shape
    for i in range(largeur):
        for j in range(hauteur):
                if nimg[i][j]>p:
                    nimg[i][j]=1
                else :
                    nimg[i][j]=0
    plt.imshow(nimg,cmap="gray")
    plt.show()
    return nimg
```

# Partie VI : Restauration d'image

```python
SP = plt.imread('images/Dead_tree_salt_and_pepper.png')
plt.imshow(SP,cmap="gray")
```

```python
from scipy import ndimage 
```

```python
def filtre_gaussian(img):
    ### filtre gaussian
    gaussian_image = ndimage.gaussian_filter( img , sigma = 7)
    plt.imshow( gaussian_image,cmap="gray" )
    plt.show()
    ### histogramme des pixels
    # image originale
    plt.hist(img.flatten(), bins=20, density=True , alpha=.5 , edgecolor='black', color='red' , label='Image originale')
    # image filtrÃ©e
    plt.hist(gaussian_image.flatten(),bins=20, density=True ,alpha=.3 ,edgecolor='black', color='blue', label='filtre gaussian')
    plt.legend()
    plt.show()
```

```python
def filtre_median(img):
    ### filtre median
    median_image = ndimage.median_filter( img , size = 8)
    plt.imshow( median_image,cmap="gray" )
    plt.show()
    ### histogramme des pixels
    # image originale
    plt.hist(img.flatten(), bins=20, density=True , alpha=.5 , edgecolor='black', color='red' , label='Image originale')
    # image filtrÃ©e
    plt.hist(median_image.flatten(), bins=20, density=True , alpha=.3 , edgecolor='black', color='blue', label='filtre median')
    plt.legend()
    plt.show()
```

```python
filtre_gaussian(SP)
```
